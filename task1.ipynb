```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('task1.csv')

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.bar(df['diagnosis'].value_counts().index, df['diagnosis'].value_counts().values, color='skyblue')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.title('Diagnosis Distribution')

plt.subplot(1, 2, 2)
plt.hist(df['radius_mean'], bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Radius Mean')
plt.ylabel('Frequency')
plt.title('Radius Mean Distribution')

plt.tight_layout()
plt.show()

```
